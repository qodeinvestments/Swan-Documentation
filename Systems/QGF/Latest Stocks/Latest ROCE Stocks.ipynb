```python
df1.dropna(subset=['QTRC_B_      Year End'],inplace=True)
df1
```

```python
df2 = df1[(df1['BDP_Mcap']>=lower_r)&(df1['BDP_Mcap']<=upper_r)]
```

```python
df2.drop(columns=['Accord Code','Sr.No.','CONS_QTRC_B_1','CONS_YR_1','CONS_QTR_B_1',"QTR_B_BS Updation Date","BDP_Close"],inplace=True)
```

**Removing the following Sectors :**

```python
removed_list = ["Finance - Investment"
                ,"Finance - NBFC"
                ,"Finance - Housing"
                ,"Construction - Real Estate"
                ,"Leather"
                ,"Bank - Public"
                ,"Aquaculture"
                ,"Breweries & Distilleries"
                ,"Bank - Private"
                ,"Finance - Others"
                ,"Finance - Asset Management"
                ,"Cigarettes/Tobacco"
                ,"Finance Term Lending"
                ,"Trading"]
```

**Finding Adjusted Profit :**  
  
  
Operating Profit Adjusted = Operating Profit *(from Ace Equity)* - Depreciation

```python
df2 = df2[~df2['CD_Industry'].isin(removed_list)]
df2['Operating Profit Adjusted'] = df2["YR_Operating Profit"]-df2["YR_Depreciation"]
df2
```

**Filter :**  
  
  
* Shareholder's Fund > 0
* Operating Profit > 0
* PAT > 0

```python
df3 = df2[(df2["QTRC_B_Shareholder's Funds"] > 0) & (df2["Operating Profit Adjusted"] > 0) & (df2["YR_Profit after tax"] > 0)]
print(df3.shape[0]," Companies present, out of ", df2.shape[0])
```

```python
# Renaming Columns for simplicity
df3.rename(columns={'QTRC_B_      Year End':'Year End',"QTRC_B_Shareholder's Funds":"Shareholder's Funds",'QTRC_B_Long Term Borrowing':"Long Term Borrowing",'QTRC_B_Short Term Borrowings':"Short Term Borrowing",
                   "YR_Operating Profit":"Operating Profit",'YR_Depreciation':"Depreciation",'YR_Profit after tax':"PAT",'YR_Other Income':"Other Income",
                   'YR_Exceptional Items':"Exceptional Items",'BDP_Mcap':"Market Cap","BDP_Close1":"Close Price"},inplace=True)
```

### Calculations :  
* **ROCE** = (Operating profit Adjusted  - Depreciation) / (shareholders fund + long term borrowings + short term borrowings)  
  
  
* **% Other Income** = Other Income / PAT
  
  
* Creating an **Alert** to know companies that have higher than 20% of *% Other Income*.

```python
df3['ROCE'] = ((df3.fillna(0)['Operating Profit Adjusted']-df3.fillna(0)['Exceptional Items']) / (df3.fillna(0)["Shareholder's Funds"]+df3.fillna(0)["Long Term Borrowing"]+df3.fillna(0)["Short Term Borrowing"]))*100
df3["% Other Income"] = (df3["Other Income"] / df3["PAT"])*100
df3["Alert"] = np.where(df3['% Other Income'] > 20,"HIGH","gtg")
df3
```

### TOP 30 Companies wo other income filter:
Finding top 30 ROCE stocks.


```python
x = df3.sort_values(by="ROCE",ascending=False)[['Company Name','Market Cap','Close Price','ROCE','Exceptional Items','% Other Income']][:30]
x.reset_index(drop=True,inplace=True)
x
```

## Final TOP 30 Companies :
Finding top 30 ROCE stocks with their % Other Income < 20 %.


```python
top30 = df3[df3['Alert']=="gtg"].sort_values(by="ROCE",ascending=False)[['Company Name','Market Cap','ROCE','Close Price']][:30]
top30.reset_index(drop=True,inplace=True)
```

## Client Side Calculations

```python
path_clients = r"C:\Users\admin\Dropbox\PMS Fundamental\2023\Ranking\PMS Clients Portfolio Holdings as of 8 June 2023.xlsx"

# Excel Sheet Names
clients = ["Vedant Ajmera", "Ashit Jhaveri", "Sanjeev Ajmera", "Shailesh Mittle",
           "Priyanka Mittle", "Tanvi Gala", "Mayuri Gandhi"]

# Cash available in each client's account
cash = {"Vedant Ajmera"     :  588316,
        "Ashit Jhaveri"     : -23081,
        "Sanjeev Ajmera"    :  1178287,
        "Shailesh Mittle"   : -124293,
        "Priyanka Mittle"   :  156408,
        "Tanvi Gala"        : -32136,
        "Mayuri Gandhi"     :  37544,}
```

```python
top30[['Qty','Qty Holding','Buy Value','Current Value']] = 0
```

```python
for r in range(top30.shape[0]):
    if top30['Company Name'].iloc[r][-1] == "." :
        top30['Company Name'].iloc[r] = top30['Company Name'].iloc[r][:-1].upper()
    else:
        top30['Company Name'].iloc[r] = top30['Company Name'].iloc[r].upper()
        
for r in range(prevholding.shape[0]):
    if prevholding['Company Name'].iloc[r][-1] == "." :
        prevholding['Company Name'].iloc[r] = prevholding['Company Name'].iloc[r][:-1].upper()
    else:
        prevholding['Company Name'].iloc[r] = prevholding['Company Name'].iloc[r].upper()
```

### Companies repeating from holdings :

```python
repeating_companies = top30[top30['Company Name'].isin(prevholding['Company Name'])].reset_index(drop=True)[['Company Name','ROCE']]

display(repeating_companies)

print(f"\nTotal {repeating_companies.shape[0]} companies are repeating from previous holdings.")
```

### Companies to go out from holdings :

```python
removing_companies = prevholding[~prevholding['Company Name'].isin(top30['Company Name'])]
removing_companies.reset_index(drop=True,inplace=True)
```

```python
display(removing_companies)

print(f"\nTotal {removing_companies.shape[0]} companies are going out from previous holdings.")
```

### Calculations :
* Number stocks to be sold.
* Amount we'll be getting after selling + available cash.
* Amount available for each new Individual stock to be bought.

```python
selling_amount = (removing_companies['Qty']*removing_companies['Close Price']).sum() + cash[clients[0]]
selling_count  = removing_companies.shape[0]
indv_amount = selling_amount/selling_count

print("Number of companies to be sold : ",selling_count,"\n")
print("Selling Amount : ",selling_amount,"\n")
print("Individual Buying Amount : ",indv_amount,"\n")
#print(selling_amount, selling_count, indv_amount)
```

### Adding Quantity to buy for new companies
* Quantity to be bought for each stock.
* Buying Value for each stock.
* Current Value of each stock.

```python
for j in range(top30.shape[0]):
    
    if top30['Company Name'].iloc[j] in list(prevholding['Company Name']):
        top30['Qty Holding'].iloc[j]   = prevholding[prevholding['Company Name']==top30['Company Name'].iloc[j]]['Qty'].iloc[0]
        top30['Close Price'].iloc[j]   = prevholding[prevholding['Company Name']==top30['Company Name'].iloc[j]]['Close Price'].iloc[0]
        top30['Current Value'].iloc[j] = top30['Qty Holding'].iloc[j]*top30['Close Price'].iloc[j]
        
    else:
        top30['Qty'].iloc[j]           = math.floor(indv_amount/top30['Close Price'].iloc[j])
        #top30['Qty Holding'].iloc[j]   = top30['Qty'].iloc[j]
        top30['Buy Value'].iloc[j]     = top30['Qty'].iloc[j]*top30['Close Price'].iloc[j]
        top30['Current Value'].iloc[j] = top30['Qty'].iloc[j]*top30['Close Price'].iloc[j]
```
