## FILTERING COMPANIES FROM YEAR 2004 ONWARDS

```python
dict_including_all = {}
market_cap = 100
temp = df_without_index
for index, row in temp.iterrows():
    dict_including_all[row[1]]={}
    flag=0
    for i in range(temp.shape[1]-1,1,-1):
        if(l[i]>=str(200403)):
            
            if(flag==0):
                if(row[i]>market_cap):
                    dict_including_all[row[1]][l[i]] = row[i]
                    flag=1
            elif(flag==1):
                if(np.isnan(row[i])==True):
                    continue
                else:
                    dict_including_all[row[1]][l[i]] = row[i]
        else:
            continue
#dict_including_all
```

```python
final_including_all = {k: v for k,v in dict_including_all.items() if v}
final_including_all
```
