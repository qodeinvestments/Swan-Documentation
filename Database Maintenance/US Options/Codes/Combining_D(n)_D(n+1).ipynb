```python
import pandas as pd
import numpy as np
```

```python
file_path1 = rf"C:\Users\admin\Desktop\USOptions\DTE0.csv"
file_path2 = rf"C:\Users\admin\Desktop\USOptions\DTE1.csv"
output_path = rf'C:\\users\\admin\\desktop\\'

df1 = pd.read_csv(file_path1)
print(df1.shape[0])
df1['Date'] = pd.to_datetime(df1['Date'],format='mixed',dayfirst=True)
df1['Expiry_Dates'] = pd.to_datetime(df1['Expiry_Dates'],format='mixed',dayfirst=True)
df1 = df1.sort_values(by=['Date','Time']).reset_index(drop=True)
df1['Option_Type'] = df1.loc[:,'Ticker'].str[-9]

df2 = pd.read_csv(file_path2)
df2['Date'] = pd.to_datetime(df2['Date'],format='mixed',dayfirst=True)
df2['Expiry_Dates'] = pd.to_datetime(df2['Expiry_Dates'],format='mixed',dayfirst=True)
df2 = df2.sort_values(by=['Date','Time']).reset_index(drop=True)
df2['Option_Type'] = df2.loc[:,'Ticker'].str[-9]

df3 = df1.merge(df2[['Date','Time','Strike','close','Option_Type']], on =['Date','Time','Strike','Option_Type'],how='left')
df3.rename(columns={'close_x':'close','close_y':'AUX_Close'},inplace=True)
df3 = df3[['New_Ticker','Date','Time','open','high','low','close','volume','AUX_Close']]
df3.to_csv(output_path + 'Combined.csv',index=False)
df3
```
